```python
## Section 03: Replacing 'NaN' valued data with the mean of its feature
```

```python
#df.isnull()
#df.mean()
df.fillna(df.mean())
df.head()
```

```python
# data shuffling
df = df.sample(frac = 1)
df.head()
```

```python
# label encoding
dfNew = df.drop(['label','feature2'], axis=1)
label = df['label']
for c in range(len(label)):
    if label.iloc[c] == "g":
        label.iloc[c] = 1
    else:
        label.iloc[c] = 0
dfNew.head()
#print(label)
```

```python
for c in dfNew.dtypes:
    print(c)
```

```python
dfNew = dfNew.astype('float64')
```

```python
## Section 03 Ends Here
```

```python
## Standardize the input data
```

```python
## section 04: Splitting data into 50% train, 30% validate and 20% test
```

```python
# Applying Train Test Split
X_train, X_test, y_train, y_test = train_test_split(dfNew, label, test_size=0.4, random_state=42)
#X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.29, random_state=42)
```

```python
## Standardize the input data
#print(Y)
mean = X_train.mean(axis=0)
std = X_train.std(axis=0)
X_train -= mean
X_train /= std

X_test -= mean
X_test /= std

#X.head()
#Y.head()
```

```python
## Section 05: Building Model
print(X_train.shape[1])
```

```python
def build_model():
    model = models.Sequential()
    model.add(layers.Dense(1, activation='relu', input_shape=(X_train.shape[1],)))
    model.add(layers.Dense(16, activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))
    model.compile(optimizer=optimizers.RMSprop(lr=0.001), loss=losses.binary_crossentropy, metrics=[metrics.binary_accuracy])
    return model
```

```python
## End of Section 05
```

```python
## Section 06: Model Fitting
```

```python
mdl = build_model()
#[print(i.shape, i.dtype) for i in model.inputs]
#[print(o.shape, o.dtype) for o in model.outputs]
#[print(l.name, l.input_shape, l.dtype) for l in model.layers]
```

```python
#mdl.fit(X_train, y_train, epochs=100, batch_size=1, verbose=2,validation_data=(X_val,y_val))
y_train = tf.convert_to_tensor(y_train, dtype=tf.float32)
mdl.fit(X_train, y_train, epochs=100, batch_size=1, verbose=2)
```

```python
y_test = tf.convert_to_tensor(y_test, dtype=tf.float32)
mdl.evaluate(X_test,y_test)
```
