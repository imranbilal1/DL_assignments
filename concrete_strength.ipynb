```python
## Standardize the input data
# Separating Independent and Dependent variables
X=df.drop(['Concrete compressive strength(MPa, megapascals) '],axis=1)
Y=df['Concrete compressive strength(MPa, megapascals) ']
#print(Y)
mean = X.mean(axis=0)
std = X.std(axis=0)
X -= mean
X /= std
print(X.head())
Y.head()
#X.head()
#Y.head()
```

```python
## section 04: Splitting data into 50% train, 30% validate and 20% test
```

```python
# Applying Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.29, random_state=42)
```

```python
## Section 05: Building Model
```

```python
def build_model():
    model = models.Sequential()
    model.add(layers.Dense(10, activation='relu',
    input_shape=(X_train.shape[1],)))
    #model.add(layers.Dropout(0.4))
    model.add(layers.Dense(8, activation='relu'))
    #model.add(layers.Dropout(0.4))
    model.add(layers.Dense(6, activation='relu'))
    #model.add(layers.Dropout(0.4))
    model.add(layers.Dense(1, activation='relu'))
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    return model
```

```python
## End of Section 05
```

```python
## Section 06: Model Fitting
```

```python
mdl = build_model()
#[print(i.shape, i.dtype) for i in model.inputs]
#[print(o.shape, o.dtype) for o in model.outputs]
#[print(l.name, l.input_shape, l.dtype) for l in model.layers]
```

```python
mdl.fit(X_train, y_train, epochs=200, batch_size=1, verbose=2,validation_data=(X_val,y_val))
```

```python
mdl.evaluate(X_test,y_test)
```
